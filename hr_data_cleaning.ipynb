```python
def parse_data(date):
    parses_data = []
    for birthdate in date:
        if re.search(r'\d{2}-\d{2}-\d{4}', birthdate) or re.search(r'\d{2}-\d{2}-\d{2}', birthdate):
            parses_data.append(datetime.strptime(birthdate, '%m-%d-%y').strftime('%Y-%m-%d'))
        elif re.search(r'\d{1,2}/\d{2}/\d{4}', birthdate):
            parses_data.append(datetime.strptime(birthdate, '%m/%d/%Y').strftime('%Y-%m-%d'))

        else:
            parses_data.append(None)
    return parses_data
```

```python
df['hire_date'] = parse_data(df['hire_date'])
df['birthdate'] = parse_data(df['birthdate'])
```

```python
def parse_termdate(date):
    parsed_termdate = []
    for term_date in date:
        if pd.isnull(term_date) or term_date.strip() == '':
            parsed_termdate.append(None)
        else:
            try:
                parsed_termdate.append(datetime.strptime(term_date, '%Y-%m-%d %H:%M:%S UTC').date().strftime('%Y-%m-%d'))
            except ValueError:
                try:
                    parsed_termdate.append(datetime.strptime(term_date, '%m/%d/%Y').date().strftime('%Y-%m-%d'))
                except ValueError:
                    parsed_termdate.append(None)
    return parsed_termdate
```

```python
df['termdate'] = parse_termdate(df['termdate'])
```
